```python
marge = 10



kanjis_interet = set(kanjis_focus)
list_kanjis_connus = set(kanjis_connus)
def filtrer_phrases(phrase):
    contient_kanji_interet = any(kanji in kanjis_interet for kanji in phrase)
    tous_kanjis_connus = all(kanji in list_kanjis_connus.union(kanjis_interet) for kanji in phrase)
    return contient_kanji_interet and tous_kanjis_connus

df_filtre = all_df[all_df['Modifiée'].apply(filtrer_phrases)]

def filtrer_phrases_margin(phrase):
    contient_kanji_interet = any(kanji in kanjis_interet for kanji in phrase)
    tous_kanjis_connus = all(kanji in list_kanjis_connus.union(kanjis_interet) for kanji in phrase)
    kanjis_inconnus = [kanji for kanji in phrase if kanji not in list_kanjis_connus.union(kanjis_interet)]
    autoriser_kanjis_inconnus = len(kanjis_inconnus) <= marge
    return contient_kanji_interet and autoriser_kanjis_inconnus

df_filtre_margin = all_df[all_df['Modifiée'].apply(filtrer_phrases_margin)]
```
